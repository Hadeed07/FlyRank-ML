## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

What decision does your work improve?\
Whether a content team should proactively intervene on a piece of content — refresh it, promote it, or leave it alone — before its performance actually moves. Without your model, they only find out a page declined after it's already declined (reactive).

Who acts on it?\
Realistically, a content or SEO team member reviewing a prioritized list — not an automated system. This matters a lot for scope: your output isn't "the model decides," it's "the model ranks candidates for a human to check."


What does a wrong recommendation cost?\
Two different costs, and they're asymmetric:\
False positive (flagged as declining, but it wasn't going to): wasted review time — someone checks a page that didn't need it. Annoying, low-cost.
False negative (a page silently declines and your model missed it): the team never sees it, so nothing gets fixed. This is the more expensive failure, since it means the model failed at its actual job with no fallback.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

Observed:<br> "Content with [X characteristic] showed [Y trend] over the observed 30/90-day window." These are descriptions of what happened in this dataset, not universal laws.


Directional:<br> "Pages with lower freshness scores and higher content age were disproportionately represented among decliners." This tells you the direction of association, and roughly how strong it is — not a guarantee for any individual page.


Decision-support:<br> "This model ranks pages by estimated risk of decline, so a content team can prioritize review — it does not replace human judgment, it narrows where a human should look first." The output is a ranked shortlist for review, not an automated verdict.
